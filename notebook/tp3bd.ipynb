# Tarefa 6
Cada relação é dividida em páginas de tamanho fixo (geralmente 8 KB). Cada página contém um cabeçalho, dados da tupla e informações de controle.
O tamanho da página no PostgreSQL é uma configuração definida durante a criação do banco de dados. O tamanho padrão é 8 KB, mas também é possível usar páginas de 4 KB, 16 KB ou 32 KB, dependendo da necessidade de desempenho e da quantidade de dados.  Quando o tamanho dos arquivos do banco de dados atinge um limite pré-definido (geralmente 1 GB), o PostgreSQL cria uma extensão para o banco de dados, adicionando um novo conjunto de arquivos para armazenar dados adicionais.O PostgreSQL utiliza um log de transações (WAL - Write-Ahead Log) para garantir a durabilidade dos dados. O WAL registra todas as alterações nos dados antes de efetivamente gravá-las nos arquivos do banco de dados.

# Tarefa 7

Ext2 (Second Extended File System):

Descrição: O Ext2 foi o sistema de arquivos padrão do Linux por muitos anos. É conhecido por sua simplicidade e confiabilidade, mas não possui recursos avançados de journaling.

Principais Características:

Não possui journaling, o que pode resultar em perda de dados em caso de falhas de energia ou travamentos do sistema.

Suporta tamanhos de arquivo de até 2 terabytes.

Suporta tamanhos de bloco de 1 KB a 4 KB.

Não possui suporte para snapshots.

Não suporta compressão de dados.

Não suporta criptografia.

Ext3 (Third Extended File System):

Descrição: O Ext3 é uma extensão do Ext2 que adiciona recursos de journaling, fornecendo maior segurança e recuperação de dados em caso de falhas do sistema.

Principais Características:

Utiliza journaling, o que reduz o tempo de recuperação do sistema em caso de falhas.

Suporta tamanhos de arquivo de até 2 terabytes.

Suporta tamanhos de bloco de 1 KB a 4 KB.

Não possui suporte nativo para snapshots, mas pode ser combinado com outras ferramentas para esse fim.

Não suporta compressão de dados.

Não suporta criptografia.

ReiserFS:

Descrição: O ReiserFS é um sistema de arquivos projetado para oferecer um desempenho superior em relação a outros sistemas de arquivos, especialmente para pequenos arquivos.

Principais Características:

Utiliza journaling para recuperação de dados em caso de falhas.

Suporta tamanhos de arquivo de até 8 terabytes.

Suporta tamanhos de bloco variáveis, otimizando o desempenho para diferentes tipos de arquivos.

Possui suporte nativo para snapshots, permitindo a criação de pontos de restauração.

Não suporta compressão de dados.

Não suporta criptografia.

XFS:

Descrição: O XFS foi projetado para lidar com grandes volumes de dados e arquivos, com foco em desempenho e escalabilidade.

Principais Características:

Utiliza journaling para recuperação de dados em caso de falhas.

Suporta tamanhos de arquivo de até 8 exabytes.

Suporta tamanhos de bloco de 512 bytes a 64 KB.

Possui suporte nativo para snapshots.

Suporta compressão de dados.

Possui suporte nativo para criptografia.
